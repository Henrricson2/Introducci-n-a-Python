# Trabajo

En este proyecto, los estudiantes desarrollarán un sistema completo en Python que integra varias de las habilidades fundamentales vistas en clase. La idea es construir un programa interactivo que permita registrar información, procesarla, analizar datos numéricos y realizar simulaciones, todo dentro de una interfaz basada en un menú.

Este trabajo busca que el estudiante combine de manera práctica y aplicada:

- Entrada y validación de datos

- Uso de condicionales y ciclos

- Manejo de listas, diccionarios y tuplas

- Procesamiento y manipulación de textos

- Operaciones matemáticas básicas

- Uso de módulos estándar de Python (especialmente random)

- Organización del programa en funciones


# Objetivo general

Diseñar un programa en Python que funcione como un Sistema Integral de Gestión, capaz de registrar usuarios, analizar textos, procesar listas numéricas y generar simulaciones aleatorias, mostrando estadísticas del funcionamiento del sistema.

# Descripción de las funcionalidades del sistema

El sistema cuenta con un menú principal interactivo, desde el cual el usuario puede acceder a los siguientes módulos:

## 1. Registro de Usuarios

El sistema solicita al usuario información personal:

- Nombre completo

- Edad

- Ciudad

- Profesión

# 2. Procesamiento de Texto

El usuario ingresa una frase y el sistema debe mostrar:

- Total de caracteres

- Número de palabras

- La frase completamente en mayúsculas

- Cantidad de apariciones de cada vocal

- La frase invertida




## 3. Registro y Análisis de Números

Se solicita al usuario ingresar una cantidad N de números.
Posteriormente, el sistema calcula:

- Promedio

- Número mayor y menor

- Suma total

- Una lista formada únicamente por los números pares


##  4. Simulación Aleatoria con random

El sistema genera automáticamente 10 números aleatorios entre 1 y 100 y:

- Los almacena en una lista

- Muestra dicha lista

- Calcula cuántas veces aparece cada número (con un diccionario)

- Indica el número más repetido



## 5. Estadísticas del Sistema

El programa llevará conteo de:

Usuarios registrados

Frases analizadas

Listas numéricas procesadas

Simulaciones realizadas
