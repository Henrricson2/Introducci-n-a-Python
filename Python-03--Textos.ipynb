## Acomodando texto en un formato

```python
ancho_línea = 54

print("=" * ancho_línea)
print("Título de mi libro".upper().center(ancho_línea))
print("Fulano de la O".center(ancho_línea))
print("\n")



print("EC4301".rjust(ancho_línea))
print("\n")
print("/°\\" * (ancho_línea//3))
print("\\_/" * (ancho_línea//3))
```
